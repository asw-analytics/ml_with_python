```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

---
## Class Sigmoid

```python
class Sigmoid():

    def __init__(self):
        '''Class Instance Constructor'''
        pass
    
    def sigmoid_method(self, z):
        '''Sigmoid Function
            
            Parameter:
            ---------
            z: np.ndarray
        '''
        return 1.0 / (1.0 + np.exp(-z))
```

---
## Construct Values

```python
# retrieve evenly spaced values between -7 and +7 with 0.1 steps --> we will receive the Sigmoid-Function for values between -7 and +7
z = np.arange(-7, 7, 0.1)
z
```

```python
phi_z = Sigmoid().sigmoid_method(z=z)
phi_z
```

---
## PLOT RESULT

```python
figure(figsize=(12, 8), dpi=80)
plt.plot(z, phi_z, linewidth=5, color='hotpink')
plt.axvline(0.0, color='g', linewidth=3)
plt.ylim(-0.1, +1.1)   # Ranges to visualize the Y-Axis
plt.xlabel('z', fontdict={'size':30, 'color': 'blue'})
plt.ylabel('$\phi (z)$', fontdict={'size':30, 'color': 'blue'})
plt.title("Sigmoid-Function", fontdict={'size':30, 'color': 'blue'})

# y-Axis marks and raster
plt.yticks([0.0, 0.25, 0.5, .75, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)
plt.tight_layout()
plt.show()
```

## STRAFFUNKTION ALS GRADIENTENABSTIEG PLOTTEN
Bei der logistischen Regression (Klassifizierungsalgorithmus) haben wir eine Likelyhood-Methode, weil es sich um eine 
Wahrscheinlichkeitsrechnung handelt. AnnÃ¤herung an >= 0.5 = 1 <= 0.5 = 0. 

```python
def cost_1(z):
    return - np.log(Sigmoid().sigmoid_method(z))

def cost_0(z):
    return - np.log(1 - Sigmoid().sigmoid_method(z))

figure(figsize=(8, 5), dpi=150)
z = np.arange(-10, 10, 0.1)
phi_z = Sigmoid().sigmoid_method(z=z)
c1 = [cost_1(x) for x in z]
plt.plot(phi_z, c1, label="J(w), wenn y=1", linewidth=3, color="darkblue")
c0 = [cost_0(x) for x in z]
plt.plot(phi_z, c0, label="J(w), wenn y=0", linestyle="--", linewidth=3, color="hotpink")
plt.ylim(0.0, 8.0)
plt.xlim([0, 1])
plt.xlabel("$\phi$(z)")
plt.ylabel("J(w)")
plt.legend(loc="best")
plt.grid()
plt.title("Gradient Descent for Logistic Regression Classification", fontdict={"color":"blue", "size":20})
plt.tight_layout()
plt.show()
```
