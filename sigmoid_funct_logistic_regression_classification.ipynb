```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

---
## Class Sigmoid

```python
class Sigmoid():

    def __init__(self):
        '''Class Instance Constructor'''
        pass
    
    def sigmoid_method(self, z):
        '''Sigmoid Function
            
            Parameter:
            ---------
            z: np.ndarray
        '''
        return 1.0 / (1.0 + np.exp(-z))
```

---
## Construct Values

```python
# retrieve evenly spaced values between -7 and +7 with 0.1 steps --> we will receive the Sigmoid-Function for values between -7 and +7
z = np.arange(-7, 7, 0.1)
z
```

```python
phi_z = Sigmoid().sigmoid_method(z=z)
phi_z
```

---
## PLOT RESULT

```python
figure(figsize=(12, 8), dpi=80)
plt.plot(z, phi_z, linewidth=5, color='hotpink')
plt.axvline(0.0, color='g', linewidth=3)
plt.ylim(-0.1, +1.1)   # Ranges to visualize the Y-Axis
plt.xlabel('z', fontdict={'size':30, 'color': 'blue'})
plt.ylabel('$\phi (z)$', fontdict={'size':30, 'color': 'blue'})
plt.title("Sigmoid-Function", fontdict={'size':30, 'color': 'blue'})

# y-Axis marks and raster
plt.yticks([0.0, 0.25, 0.5, .75, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)
plt.tight_layout()
plt.show()
```
